```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to find all CSV files in a directory and its subdirectories
def find_csv_files(directory):
    csv_files = []
    csv_file_names = []

    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".CSV"):
                csv_files.append(os.path.join(root, file))
                csv_file_names.append(file)

    return csv_files, csv_file_names

search_directory = "D:/Versuch 41"

csv_file_paths, csv_file_names = find_csv_files(search_directory)


dataframes = [pd.read_csv(csv_path) for csv_path in csv_file_paths]
```

```python
dataframes = []
for csv_path in csv_file_paths:
    df = pd.read_csv(csv_path)
    dataframes.append(df)
   
```

```python
count = 0
for df in dataframes:
    #meta_data
    meta_data = []
    meta_list = ['Vertical Units', 'Vertical Scale', 'Horizontal Units', 'Horizontal Scale']
    for element in meta_list:
        row = df.loc[df['Record Length'] == element]
        meta_data.append(row.iloc[0, 1])
    meta_data[1].split('.')[0]

    #actual_data
    df_data = df
    df_data.head()
    x =  int(meta_data[3].split('.')[0])*df.iloc[:, 3]
    y =  int(meta_data[1].split('.')[0])*df.iloc[:, 4]
    plt.plot(x,y, label = 'Reihe 1', c = 'firebrick')
    plt.title(csv_file_names[count])
    plt.xlabel('Zeit in '+str(meta_data[2]))
    plt.ylabel('gemessene Spannung in '+str(meta_data[0]))
    plt.legend()
    fig = plt.gcf()
    fig.set_size_inches(8, 6) 

    plt.savefig('images'+csv_file_names[count]+'.png', dpi=500)
    plt.clf()
    count = count+1
```
